```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = {
    "Product": ["Laptop","Mouse","Desk Chair","Monitor","Notebook","Tablet","Desk","Keyboard","Printer","Laptop"],
    "Category": ["Electronics","Electronics","Furniture","Electronics","Office Supplies","Electronics","Furniture","Electronics","Office Supplies","Electronics"],
    "Region": ["Europe","Europe","Asia","Asia","Europe","North America","North America","Europe","Asia","North America"],
    "Quantity": [1,3,2,1,5,2,1,4,1,1],
    "Unit Price": [1200,25,150,300,5,400,500,45,200,1200]
}

df = pd.DataFrame(data)
df
```

```python
df["Revenue"] = df["Quantity"] * df["Unit Price"]
df
```

```python
df["Revenue"].sum()
```

```python
df.groupby("Region")["Revenue"].sum()
```

```python
df.groupby("Category")["Revenue"].sum()
```

```python
df.groupby("Region")["Revenue"].sum().plot(kind="bar")

plt.title("Revenue by Region")
plt.ylabel("Revenue")
plt.show()
```

```python
df.groupby("Category")["Revenue"].sum().plot(kind="bar")

plt.title("Revenue by Category")
plt.ylabel("Revenue")
plt.show()
```
